# 01 - Data Cleaning
This notebook performs data cleaning and preprocessing for subscription data.

```

import pandas as pd

# Load dataset
df = pd.read_csv('../data/raw/subscription_data.csv')

# Remove duplicates
df = df.drop_duplicates(subset='customer_id')

# Convert date columns
df['signup_date'] = pd.to_datetime(df['signup_date'])
df['churn_date'] = pd.to_datetime(df['churn_date'])

# Create churn flag
df['churn_flag'] = df['churn_date'].notna().astype(int)

# Save cleaned dataset
df.to_csv('../data/processed/cleaned_subscription_data.csv', index=False)

df.head()
```
