# 03 - Cohort Analysis
This notebook performs signup cohort retention analysis.

```

import pandas as pd

df = pd.read_csv('../data/processed/cleaned_subscription_data.csv')
df['signup_month'] = pd.to_datetime(df['signup_date']).dt.to_period('M')

cohort_counts = df.groupby('signup_month')['customer_id'].nunique()

print(cohort_counts)
```
