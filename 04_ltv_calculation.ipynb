# 04 - Customer Lifetime Value (LTV)
This notebook calculates ARPU and LTV.

```

import pandas as pd

df = pd.read_csv('../data/processed/cleaned_subscription_data.csv')

arpu = df['monthly_revenue'].mean()
monthly_churn = df['churn_flag'].mean()

ltv = arpu / monthly_churn if monthly_churn != 0 else 0

print("ARPU:", round(arpu,2))
print("Monthly Churn Rate:", round(monthly_churn*100,2), "%")
print("Estimated LTV:", round(ltv,2))
```
